## **Pandas : What is a DataFrame?**

---------

### **DataFrame = Multiple Series Aligned on the Same Index**

A DataFrame is a collection of Series that all share the same index. Each column is one Series. Every row is one observation.

```
Python dict of lists            Pandas DataFrame
──────────────────              ─────────────────
{"Name": ["Ash","Misty"],       Name   HP   Type
 "HP":   [45, 60],          →   Ash    45   Fire
 "Type": ["Fire","Water"]}      Misty  60   Water
```

-   **Rows**: one observation per row — one Pokemon, one sensor reading, one sale.

-   **Columns**: one variable per column — each is a Series with the same index.

-   **Index**: shared row labels — default is 0, 1, 2... but can be set to any meaningful label.

> **Key insight:** A DataFrame is a dict of aligned Series. Every Series skill you learned in Week 1 works on any DataFrame column.

## **Pandas : Creating a DataFrame**

---------

You create a DataFrame using `pd.DataFrame()`. The most common starting point is a **dictionary of lists** — each key becomes a column name and each list becomes the column data.

-   **`data`**: A dict of lists, a list of dicts, or a NumPy array.

-   **`columns`**: Override the column names.

-   **`index`**: Set custom row labels instead of the default 0, 1, 2...

````python
# --- From a dictionary of lists ---
# Each key becomes a column, each list becomes the column data
pokemon = pd.DataFrame(
    {
        "Name": ["Bulbasaur", "Charmander", "Squirtle", "Pikachu", "Mewtwo"],
        "Type": ["Grass", "Fire", "Water", "Electric", "Psychic"],
        "HP": [45, 39, 44, 35, 106],
        "Attack": [49, 52, 48, 55, 110],
    }
)
print(pokemon)
````

````python
# --- From a list of dicts ---
# Each dict is one row
rows = [
    {"City": "Istanbul", "Country": "Turkey", "Pop_M": 15.8},
    {"City": "Ankara", "Country": "Turkey", "Pop_M": 5.7},
    {"City": "Izmir", "Country": "Turkey", "Pop_M": 4.4},
]
cities = pd.DataFrame(rows)
print(cities)
````

## **Pandas : Core Attributes**

---------

DataFrames share the same attribute pattern as Series — no parentheses.

-   **`.shape`**: A tuple `(rows, columns)` — the first thing to check on any new dataset.

-   **`.columns`**: The column names as an Index object.

-   **`.index`**: The row labels.

-   **`.dtypes`**: The data type of every column.

-   **`.size`**: Total number of cells (rows x columns).

> **First thing to check on a new DataFrame:** `.shape` tells you how big the data is before you do anything else.

````python
# --- Core Attributes ---
print(pokemon.shape)  # (5, 4)  ← 5 rows, 4 columns
print(pokemon.columns)  # Index(['Name', 'Type', 'HP', 'Attack'])
print(pokemon.index)  # RangeIndex(start=0, stop=5, step=1)
print(pokemon.dtypes)  # Name: object, Type: object, HP: int64, Attack: int64
print(pokemon.size)  # 20  ← 5 rows x 4 columns
````

## **Pandas : Reading from CSV — pd.read_csv()**

---------

`pd.read_csv()` is the standard way to load data from a file. Key parameters:

-   **`filepath`**: Path to the CSV file.

-   **`index_col`**: Column to use as the row index instead of 0, 1, 2...

-   **`usecols`**: A list of column names to load — skip the rest.

-   **`nrows`**: Load only the first n rows — useful for previewing huge files.

-   **`dtype`**: Force specific column types on load.

> On large files always use `nrows=100` first to confirm the structure before loading everything.

````python
# --- Create a sample CSV to work with ---
import os

os.makedirs("data", exist_ok=True)

sample = pd.DataFrame(
    {
        "Name": [
            "Bulbasaur",
            "Charmander",
            "Squirtle",
            "Pikachu",
            "Mewtwo",
            "Eevee",
            "Snorlax",
            "Gengar",
            "Machamp",
            "Gyarados",
        ],
        "Type 1": [
            "Grass",
            "Fire",
            "Water",
            "Electric",
            "Psychic",
            "Normal",
            "Normal",
            "Ghost",
            "Fighting",
            "Water",
        ],
        "HP": [45, 39, 44, 35, 106, 55, 160, 60, 80, 95],
        "Attack": [49, 52, 48, 55, 110, 55, 110, 65, 130, 125],
        "Defense": [49, 43, 65, 40, 90, 50, 65, 60, 80, 79],
        "Speed": [45, 65, 43, 90, 130, 55, 30, 110, 55, 81],
        "Legendary": [
            False,
            False,
            False,
            False,
            True,
            False,
            False,
            False,
            False,
            False,
        ],
    }
)
sample.to_csv("data/pokemon.csv", index=False)
print("pokemon.csv created ✓")
````

````python
# --- Load the full CSV ---
df = pd.read_csv("data/pokemon.csv")
print(df)

print("\n")  # --- IGNORE ---

# --- Load only specific columns ---
df_slim = pd.read_csv("data/pokemon.csv", usecols=["Name", "Type 1", "HP"])
print(df_slim)
````

## **Pandas : Inspecting — .info() and .describe()**

---------

Two methods give you a full picture of any DataFrame in seconds:

-   **`.info()`**: Prints shape, column names, non-null counts, dtypes, and memory usage. Run this first on every new dataset — it reveals missing values and type problems immediately.

-   **`.describe()`**: Returns summary statistics for all numeric columns: count, mean, std, min, 25th/50th/75th percentile, max.

> **`.info()` is the most important method in this session.** Every line tells you something: a column where non-null count is less than the total rows has missing values.

````python
# --- .info() : full structural overview ---
# Read every line:
# RangeIndex: how many rows
# Non-Null Count: if less than total rows, there are missing values
# Dtype: object = text, int64/float64 = numeric, bool = True/False
df.info()
````

````python
# --- .describe() : statistical summary of numeric columns ---
# count: how many non-null values
# mean:  average
# std:   standard deviation (spread)
# min/max: smallest and largest value
# 25%/50%/75%: quartiles
print(df.describe())
````

````python
# --- .describe() on non-numeric columns ---
# include="object" shows text column stats: count, unique, top, freq
print(df.describe(include="object"))
````

## **Pandas : Previewing — .head(), .tail(), .sample()**

---------

Three methods let you peek at rows without printing the entire DataFrame:

-   **`.head(n)`**: Returns the first `n` rows. Default is 5.

-   **`.tail(n)`**: Returns the last `n` rows. Default is 5.

-   **`.sample(n)`**: Returns `n` randomly selected rows. Useful on large datasets where the first and last rows are not representative.

> Always use `.sample()` in addition to `.head()` on large datasets. The first rows are often the "clean" rows — problems hide further down.

````python
# --- Preview rows ---
print(df.head(3))  # first 3 rows

print("\n")  # --- IGNORE ---

print(df.tail(2))  # last 2 rows

print("\n")  # --- IGNORE ---

print(df.sample(3))  # 3 random rows (different each run)
````

## **Pandas : Accessing Columns**

---------

There are two ways to select a column from a DataFrame:

-   **`df["column"]`**: Returns a **Series**. Always works. Use for column names with spaces or that match method names.

-   **`df.column`**: Dot notation. Works only when the column name has no spaces and does not clash with a DataFrame method (e.g. avoid `df.mean` as a column name).

-   **`df[["col1", "col2"]]`**: Pass a **list** inside `[]` to select multiple columns — returns a **DataFrame**, not a Series.

> **Prefer `df["column"]` over `df.column`.** Dot notation breaks for column names with spaces, special characters, or names that clash with Pandas methods. Building the habit early prevents bugs later.

````python
# --- Single column → returns a Series ---
print(df["HP"])  # bracket notation  ← always safe

print("\n")  # --- IGNORE ---

print(type(df["HP"]))  # <class 'pandas.core.series.Series'>
````

````python
# --- Multiple columns → returns a DataFrame ---
# Note: double brackets [[]] are required
subset = df[["Name", "HP", "Attack"]]
print(subset)

print("\n")  # --- IGNORE ---

print(type(subset))  # <class 'pandas.core.frame.DataFrame'>
````

## **Pandas : .value_counts() on a DataFrame Column**

---------

`.value_counts()` works exactly the same on a DataFrame column as it did on a standalone Series. Select a column with `df["col"]`, then chain `.value_counts()`.

-   **Default**: sorted by frequency, highest first.

-   **`normalize=True`**: Returns proportions (0.0 to 1.0) instead of counts.

> This is the **first method to run on any categorical column** of a DataFrame. Run it immediately after `.info()` to understand the distribution of your key grouping variables.

````python
# --- .value_counts() on a DataFrame column ---
print(df["Type 1"].value_counts())

print("\n")  # --- IGNORE ---

# --- As percentages ---
print(df["Type 1"].value_counts(normalize=True).round(2))

print("\n")  # --- IGNORE ---

# --- Boolean column: proportion of Legendary Pokemon ---
print(df["Legendary"].value_counts(normalize=True))
````
