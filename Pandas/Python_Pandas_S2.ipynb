## **Pandas : Boolean Filtering**

---------

Boolean filtering lets you ask a question about every element and keep only those that answer `True`. It follows a two-step pattern:

-   **Step 1 — Create a mask**: Apply a condition to produce a Series of `True` / `False` values.

-   **Step 2 — Apply the mask**: Pass it inside `[]` to filter the original Series.

> When combining conditions, always wrap each condition in parentheses and use `&` (AND), `|` (OR), or `~` (NOT) — never Python's `and` / `or`.

````python
# --- Setup ---
weather = pd.Series(
    {"Mon": 30, "Tue": 32, "Wed": 28, "Thu": 35},
    name="Temp (C)"
)

# --- Step 1 : create the mask ---
is_hot = weather > 30
print(is_hot)   # True/False for every element
````

````python
# --- Step 2 : apply the mask ---
print(weather[is_hot])   # days where temp > 30

print("\n") # --- IGNORE ---

# --- One-liner (most common style) ---
print(weather[weather > 30])
````

## **Pandas : .between() and .isin()**

---------

Two shortcuts that replace verbose compound conditions:

-   **`.between(a, b)`**: A cleaner replacement for `(s >= a) & (s <= b)` — both ends are inclusive.

-   **`.isin([...])`**: Returns `True` wherever the value is in the provided list — useful for categorical data.

> Both methods return a boolean mask, so they are always used inside `[]` to filter.

````python
# --- .between() ---
print(weather[weather.between(30, 32)])   # same as (weather >= 30) & (weather <= 32)

print("\n") # --- IGNORE ---

# --- .isin() ---
pokemon_types = pd.Series(["Fire", "Water", "Fire", "Grass", "Water", "Electric"])

# Keep only Fire and Water types
mask = pokemon_types.isin(["Fire", "Water"])
print(pokemon_types[mask])
````

## **Pandas : Sorting**

---------

Two methods control how a Series is ordered:

-   **`.sort_values()`**: Orders by the **data values**. Use `ascending=False` to reverse.

-   **`.sort_index()`**: Orders by the **index labels** alphabetically or numerically.

-   **`.nlargest(n)`** / **`.nsmallest(n)`**: Returns the top or bottom `n` values without sorting the entire Series.

````python
# --- Sort by values ---
print(weather.sort_values())                 # coldest to hottest

print("\n") # --- IGNORE ---

print(weather.sort_values(ascending=False))  # hottest to coldest
````

````python
# --- Sort by index ---
print(weather.sort_index())   # Mon, Thu, Tue, Wed (alphabetical)

print("\n") # --- IGNORE ---

# --- Top / bottom n values ---
print(weather.nlargest(2))    # 2 hottest days

print("\n") # --- IGNORE ---

print(weather.nsmallest(2))   # 2 coldest days
````

## **Pandas : The .apply() Method**

---------

When built-in methods are not enough, `.apply()` runs **any Python function** on every element of the Series.

-   **Named function**: Define with `def`, then pass the name to `.apply()` — no parentheses.

-   **Lambda function**: An anonymous one-liner written directly inside `.apply()` — best for simple logic.

> **Performance note:** `.apply()` is much slower than vectorized operations (e.g. `s * 2`). Use it only when no built-in method can do the job.

````python
# --- Named function ---
def celsius_to_fahrenheit(c):
    return (c * 9/5) + 32

print(weather.apply(celsius_to_fahrenheit))
````

````python
# --- Lambda (one-liner, same result) ---
print(weather.apply(lambda c: (c * 9/5) + 32))

print("\n") # --- IGNORE ---

# --- Practical example : classify scores ---
grades = pd.Series([88, 55, 72, 95, 40], index=["Ana", "Ben", "Cara", "Dan", "Eve"])

def grade_label(score):
    if score >= 90:   return "A"
    elif score >= 70: return "B"
    elif score >= 55: return "C"
    else:             return "F"

print(grades.apply(grade_label))
````
