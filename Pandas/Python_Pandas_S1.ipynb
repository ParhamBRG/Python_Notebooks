## **Pandas : What is a Series?**

---------

### **Why use a Series instead of a list?**

In a standard Python list you access data only by position — `list[0]`, `list[1]`. A Pandas Series gives every value a **label** (the index), so you can access data by a meaningful name instead of just a number.

```
Python list          Pandas Series
────────────         ─────────────
[30, 32, 28, 35]     Mon    30
                     Tue    32
 index: 0  1  2  3   Wed    28
                     Thu    35
```
"Item number 2" tells you nothing. "Wednesday" tells you everything.

## **Pandas : Creating a Series**

---------

You create a Series using `pd.Series()`. There are two common starting points:

-   **From a List:** Pandas generates a default numeric index (0, 1, 2...).

-   **From a Dictionary:** The dictionary **keys** automatically become the index labels, and the **values** become the data.

The constructor also accepts two optional parameters:

-   **`index`**: A list of custom labels when creating from a list.

-   **`name`**: A string title for the Series — useful when it becomes a DataFrame column.

````python
# --- From a List (auto numeric index) ---
ice_cream = ["Chocolate", "Vanilla", "Strawberry"]
s1 = pd.Series(ice_cream)
print(s1)

print("\n")  # --- IGNORE ---

# --- From a Dictionary (keys become the index) ---
temps = {"Mon": 30, "Tue": 32, "Wed": 28, "Thu": 35}
s2 = pd.Series(temps, name="Temp (C)")
print(s2)
````

````python
# --- From a List with a custom index and name ---
fruits = ["Apple", "Orange", "Plum"]
weekdays = ["Monday", "Tuesday", "Wednesday"]

# 'data' and 'index' are parameters; 'fruits' and 'weekdays' are arguments
s3 = pd.Series(data=fruits, index=weekdays, name="Fruit Inventory")
print(s3)
````

## **Pandas : Core Attributes**

---------

Attributes give you metadata **about** your Series. Unlike methods, attributes do **not** use parentheses `()`.

-   **`.values`**: Returns the data as a NumPy array.

-   **`.index`**: Returns the index object (the labels).

-   **`.dtype`**: Returns the data type of the underlying data.

-   **`.shape`**: Returns a tuple showing the number of elements.

-   **`.name`**: Returns the name of the Series.

````python
# --- Core Attributes ---
print(s2.values)  # array([30, 32, 28, 35])
print(s2.index)  # Index(['Mon', 'Tue', 'Wed', 'Thu'])
print(s2.dtype)  # int64
print(s2.shape)  # (4,)
print(s2.name)  # Temp (C)
````

## **Pandas : Statistical Methods**

---------

Methods perform calculations on the data. Because Pandas is built on NumPy, these run extremely fast even on millions of rows.

-   **`.sum()`**: Returns the total of all numeric values.

-   **`.mean()`**: Returns the average.

-   **`.product()`**: Multiplies all values together.

-   **`.min()`** / **`.max()`**: Returns the smallest / largest value.
-   **`.std()`**: Returns the standard deviation.

````python
# --- Statistical Methods ---
numbers = pd.Series([1, 2, 3, 4])

print(numbers.sum())  # 10
print(numbers.mean())  # 2.5
print(numbers.product())  # 24
print(numbers.min())  # 1
print(numbers.max())  # 4
print(numbers.std())  # 1.29099
````

## **Pandas : Indexing with .loc and .iloc**

---------

Pandas provides two dedicated accessors for retrieving data:

-   **`.loc[]`**: Selects by **label** — use the actual index name (`"Mon"`, `"Alice"`).

-   **`.iloc[]`**: Selects by **integer position** — use the numeric position (0, 1, 2...) like a Python list.

> **Critical difference with slicing:**
> - `.loc["Mon":"Wed"]` → **includes** `"Wed"` (label-based, both ends inclusive) **Inclusive slicing**
> - `.iloc[0:3]` → **excludes** position `3` (integer-based, like Python lists) **Exclusive slicing**

````python
# --- .loc : select by LABEL ---
print(s2.loc["Mon"])  # single value  → 30

print("\n")  # --- IGNORE ---

print(s2.loc[["Mon", "Wed"]])  # list of labels → Mon and Wed

print("\n")  # --- IGNORE ---

print(s2.loc["Mon":"Wed"])  # label slice   → Mon, Tue, Wed (Wed IS included)
````

````python
# --- .iloc : select by POSITION ---
print(s2.iloc[0])  # first item  → 30

print("\n")  # --- IGNORE ---

print(s2.iloc[-1])  # last item   → 35

print("\n")  # --- IGNORE ---

print(s2.iloc[0:3])  # positions 0, 1, 2  (3 is NOT included)
````

## **Pandas : .value_counts()**

---------

`.value_counts()` is the fastest way to profile any categorical column. It returns a frequency table sorted from most common to least common.

-   **Default**: sorted by frequency, highest first.

-   **`normalize=True`**: returns proportions (0.0 to 1.0) instead of counts.

-   **`sort=False`**: keeps the original order of the index instead of sorting by frequency.

> This is the **first method to run** on any categorical column when exploring a new dataset.

````python
# --- .value_counts() ---
types = pd.Series(["Fire", "Water", "Fire", "Grass", "Water", "Fire", "Water", "Grass"])

# Default: count, sorted highest first
print(types.value_counts())

print("\n")  # --- IGNORE ---

# As proportions
print(types.value_counts(normalize=True))
````

````python
# --- sort=False keeps the original order ---
print(types.value_counts(sort=False))
````

## **Pandas : .map()**

---------

`.map()` substitutes every value in a Series using a lookup dictionary — no loop, no `.apply()` needed.

-   **`.map(dict)`**: Replaces each value using the dict as a lookup table.

-   **Unmatched keys**: Any value not found in the dict becomes `NaN`.

-   **vs `.apply()`**: Use `.map()` for simple value substitution — it is cleaner and faster than `.apply(lambda x: ...)`.

> **Rule:** If you have a lookup table (dict), use `.map()`. If you need complex logic, use `.apply()`.

````python
# --- .map() with a dictionary ---
type_codes = {"Fire": "Fi", "Water": "Wa", "Grass": "Gr"}

coded = types.map(type_codes)
print(coded)
````

````python
# --- Unmatched keys become NaN ---
partial = {"Fire": "Fi", "Water": "Wa"}  # Grass has no entry

print(types.map(partial))  # Grass entries → NaN

print("\n")  # --- IGNORE ---

# --- .map() vs .apply() ---
# Both lines produce the same result — map() is preferred for dict lookup
print(types.map(type_codes))  # clean
print(types.apply(lambda x: type_codes.get(x, x)))  # verbose
````
