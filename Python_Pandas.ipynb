## **Pandas : What is a Series?**

---------

### **Why use a Series?**

In standard Python lists, you access data by position. In Pandas, the **Index** provides context. It allows you to identify data points by meaningful labels (like Dates, Names, or ID numbers) rather than just "item number 5."

## **Pandas : Implementation**

---------

You can transform standard Python collections into Pandas Series using the `pd.Series()` constructor.

-   **From a List:** Pandas generates a default numeric index (0, 1, 2...).
    
-   **From a Dictionary:** The dictionary **keys** automatically become the index labels, and the **values** become the data.

```python
import pandas as pd

# --- From a List ---
ice_cream = ["Chocolate", "Vanilla", "Strawberry"]
s1 = pd.Series(ice_cream)
print(s1)

print("\n") # --- IGNORE ---

# --- From a Dictionary ---
web_dict = {"YouTube": "Video", "Google": "Search", "Amazon": "Shopping"}
s2 = pd.Series(web_dict)
print(s2)
```

## **Pandas - Core Attributes**

---------

Attributes provide metadata about your Series. Unlike methods, attributes **do not** use parentheses `()`.

-   **`.values`**: Returns the data as a NumPy array.
    
-   **`.index`**: Returns the index object (the labels).
    
-   **`.dtype`**: Returns the data type of the underlying data.

```python
# --- Checking Attributes ---
prices = pd.Series([1.99, 3.50, 5.00])

print(prices.values)  # [1.99, 3.5, 5.0]
print(prices.index)   # RangeIndex(start=0, stop=3, step=1)
print(prices.dtype)   # float64
```

## **Pandas - Parameters and Arguments in pd.Series()**

---------

The `pd.Series()` method has specific parameters that allow you to customize the object during creation.

-   **`data`**: The source data (list, dict, etc.).
    
-   **`index`**: A list of custom labels.
    
-   **`name`**: A string to provide a title to the Series.

```python
# --- Using Parameters ---
fruits = ["Apple", "Orange", "Plum"]
weekdays = ["Monday", "Tuesday", "Wednesday"]

# 'data' and 'index' are parameters; 'fruits' and 'weekdays' are arguments
pd.Series(data=fruits, index=weekdays, name="Fruit Inventory")
```

## **Pandas : Basic Methods**

---------

### **Statistical Methods**

Methods perform calculations on the data. Because Pandas is optimized for math, these are extremely fast.

-   **`.sum()`**: Returns the total of all numeric values.
    
-   **`.mean()`**: Returns the average of the values.
    
-   **`.product()`**: Multiplies all values together.

```python
# --- Statistical Calculations ---
numbers = pd.Series([1, 2, 3, 4])
print(numbers.sum())      # 10
print(numbers.mean())     # 2.5
print(numbers.product())  # 24
```

## **Pandas : Reading from CSV**

---------

Most data work involves reading files. The `pd.read_csv()` function is the standard for importing data.

-   **`usecols`**: A list parameter to select only specific columns.
    
-   **`.squeeze("columns")`**: Converts a single-column DataFrame into a Series.
    
-   **`.head(n)`**: Returns the first $n$ rows (default is 5).
    
-   **`.tail(n)`**: Returns the last $n$ rows.

```python
# --- Reading CSV ---
# Assuming a file 'pokemon.csv' exists
pokemon = pd.read_csv("data/pokemon.csv", usecols=["HP"]).squeeze("columns")

# --- Previewing ---
print(pokemon.head(3))  # First 3 rows
print(pokemon.tail(2))  # Last 2 rows
```
