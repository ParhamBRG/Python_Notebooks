## **Error Handling : Categorizing Error Types**
---

Before learning how to fix errors, you must understand the three main categories of errors in Python. Not all of them can be handled by code!

### **1. Syntax Errors (The "Grammar" Errors)**
**Description:** These occur when you break the rules of Python's language. Python cannot understand your instructions, so it stops **before** the code even runs.
* **Can you handle it?** No. You must fix the code manually before it will run.
* **Common Example:** Missing a colon `:`, wrong indentation, or unclosed parentheses.

```python
# ERROR: Missing a colon after the 'if' statement
if True
    print("This will cause a SyntaxError")
```

### **2. Runtime Errors**
**Description:** These occurs when your code is syntactically perfect (the "grammar" is correct), but an event happens during execution that Python doesn't know how to handle. These are formally called **Exceptions**.
* **Can you handle it?** Unlike Syntax Errors, Runtime Errors can be "caught" and managed using `try` and `except` blocks, allowing your program to keep running.

```python
# ERROR: Trying to divide by zero is mathematically impossible
numerator = 10
denominator = 0

print(numerator / denominator) 
# Results in: ZeroDivisionError
```


#### **Detailed Breakdown of Common Exceptions**
##### **A. ValueError**
* **Description:** Happens when a function receives an argument that has the right "type" but an inappropriate "value."
* **When to use it:** Most common when converting user input.
* **Example:** `int("Hello")` — Python knows how to turn strings into integers, but "Hello" has no numeric value.
##### **B. TypeError**
* **Description:** Happens when an operation is applied to an object of an inappropriate type.
* **Example:** `'5' + 10` — You cannot add a string (text) to an integer (number).

##### **C. IndexError**
* **Description:** Happens when you try to access an index that is out of the range of a list or tuple.
* **Example:** `my_list = [1, 2]` then calling `my_list[5]`.

##### **D. KeyError**
* **Description:** Happens when you try to access a dictionary key that does not exist.
* **Example:** `my_dict = {"name": "Alice"}` then calling `my_dict["age"]`.

##### **E. ZeroDivisionError**
* **Description:** Raised when the second argument of a division or modulo operation is zero.
* **Example:** `result = 100 / 0`.

##### **F. FileNotFoundError**
* **Description:** Raised when a file or directory is requested but doesn't exist.
* **Example:** `open("my_secret_file.txt")` when that file hasn't been created yet.

| Exception | Common Cause | Real-world Scenario |
| :--- | :--- | :--- |
| **`ValueError`** | A function gets the right type of data, but the "value" is inappropriate. | A user types their name (text) in an input field that expects an age (number). |
| **`TypeError`** | An operation is performed on incompatible data types. | Trying to calculate a total price by adding a string price like `"10$"` to a number `5`. |
| **`IndexError`** | Trying to access an index that does not exist in a list or tuple. | A photo gallery trying to show the 11th photo when the user only uploaded 3. |
| **`KeyError`** | Trying to access a dictionary key that hasn't been defined. | A login system looking for a `username` key in a database, but the user hasn't registered yet. |
| **`ZeroDivisionError`** | Attempting to divide a number by zero. | A banking app calculating interest where the time period or balance is mistakenly set to 0. |

### **3. Logical Errors (The "Sneaky" Errors)**

**Description:** The program runs without crashing and has no syntax errors, but the output is wrong. The logic of your code is flawed.

**Can you catch it with `try/except`?** No, because the program doesn't think anything is "wrong"—it's just doing exactly what you told it to do (even if what you told it was wrong).

**Example:** Using `+` instead of `*`, or calculating an average incorrectly.

```python
# GOAL: Calculate the total price of a $100 item with a $10 discount.
price = 100
discount = 10

# ERROR: Adding the discount instead of subtracting it
total = price + discount 

print(f"Total Price: ${total}") 
# Output: $110 (Wrong! It should be $90)
```

| Error Category | When does it happen? | Can `try/except` handle it? | Example |
| :--- | :--- | :--- | :--- |
| **Syntax** | Before the code runs | **No** | `if x = 5:` (should be `==`) |
| **Runtime** | While code is running | **Yes** | `int("Hello")` |
| **Logical** | While code is running | **No** | `area = length + width` (wrong formula) |

## **Error Handling : The `try` and `except` Block**
---

### **1. Basic Syntax**
The `try` block lets you test a block of code for errors. The `except` block handles the error.

```python
try:
    # Code that might cause an error
    number = int(input("Enter a number: "))
    print(10 / number)
except:
    # Code that runs if an error occurs
    print("Something went wrong!")
```

### **2. Catching Specific Exceptions**
It is better to catch specific errors so you don't accidentally hide other bugs. Common errors include `ValueError`, `ZeroDivisionError`, and `FileNotFoundError`.

```python
try:
    value = int(input("Enter a divisor: "))
    result = 10 / value
except ValueError:
    print("Error: You must enter a valid integer.")
except ZeroDivisionError:
    print("Error: You cannot divide by zero.")
```

## **Error Handling : The `else` Clause**
---

### **When to use it?**
Use the `else` block for code that should only run **if the `try` block succeeded** (no errors occurred). 



### **Why use it?**
It helps separate the **"risky"** code (the part that might fail) from the **"logic"** that follows. By putting the follow-up logic in the `else` block, you ensure that you aren't accidentally catching errors that might happen in the follow-up logic itself, which keeps your error handling precise.

```python
try:
    # Risky: Trying to convert input and divide
    number = int(input("Enter a number to divide 100 by: "))
    result = 100 / number
except ValueError:
    print("That's not a valid number!")
except ZeroDivisionError:
    print("You can't divide by zero!")
else:
    # Success: Logic that only runs if the try block worked
    print(f"Success! The result is {result}")
```

## **Error Handling : The `finally` Clause**
---

### **When to use it?**
Use the `finally` block for code that **must run no matter what**, regardless of whether an error occurred, was caught, or even if the program crashed.



### **Why use it?**
It is primarily used for **"Cleanup"**. For example, if your code opens a file or connects to a database, those resources must be closed to save memory and prevent data corruption. If the program crashes while reading a file, the `finally` block ensures that the file is closed properly before the program stops, acting as a "safety net."

```python
try:
    print("Attempting to open the system database...")
    # Imagine a crash happens here due to a calculation error
    result = 10 / 0 
except ZeroDivisionError:
    print("Error: Calculation failed, but we must still secure the system.")
finally:
    # This block runs every time to ensure safety and clean up resources
    print("Closing database connection...")
    print("Cleanup complete. System is safe.")
```

## **Error Handling Summary Table**
---

| Block | Purpose | Run Condition |
| :--- | :--- | :--- |
| **`try`** | The "Risky" Code | Always runs first to test for errors. |
| **`except`** | Error Handling | Runs only if an error occurs in the `try` block. |
| **`else`** | Success Logic | Runs only if **no** errors occur in the `try` block. |
| **`finally`** | Cleanup | Always runs at the end, regardless of errors or crashes. |

### **The Full Workflow Example**

```python
try:
    # 1. Try to do something risky
    data = int(input("Enter a number: "))
    result = 10 / data
except (ValueError, ZeroDivisionError) as e:
    # 2. Handle specific errors
    print(f"Error caught: {e}")
else:
    # 3. Only runs if the math above worked perfectly
    print(f"Calculation successful! Result: {result}")
finally:
    # 4. Always runs to finish the process
    print("Process complete.")
```

## **Error Handling : The `raise` Keyword**
---

### **What is it?**
Sometimes, a piece of code is mathematically or syntactically correct, but it violates your **business logic**. The `raise` keyword allows you to manually trigger an exception to stop the program from continuing with "bad" data.

### **Why use it?**
It is used to enforce rules. For example, if a user enters a negative age or a password that is too short, Python won't see that as a "Runtime Error," but **you** know it is wrong. By using `raise`, you can force the program into an `except` block.

```python
### **Code Example: Enforcing Age Limits**

def check_age(age):
    if age < 0:
        # Manually triggering a ValueError with a custom message
        raise ValueError("Age cannot be a negative number!")
    elif age < 18:
        print("Access Denied: You must be 18 or older.")
    else:
        print("Access Granted!")

try:
    user_age = int(input("Enter your age: "))
    check_age(user_age)
except ValueError as e:
    # This catches both conversion errors AND our raised error
    print(f"Validation Error: {e}")
```

### **How it works in the "Try-Except" flow**
1.  **`raise`** sends a signal that an error has occurred.
2.  Python immediately stops the `try` block.
3.  Python looks for a matching **`except`** block to handle the message.

```python
try:
    username = input("Create a username: ")
    if len(username) < 5:
        raise NameError("Username is too short!")
except NameError as error:
    print(f"Login Error: {error}")
```
